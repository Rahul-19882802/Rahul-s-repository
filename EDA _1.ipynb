```python
def Outlier_detection(data, colnames):
    for colname in colnames:
        q1 = data[colname].quantile(0.25)
        q3 = data[colname].quantile(0.75)
        iqr = q3 - q1
        upper_extreme = q3 + (1.5 * iqr)
        lower_extreme = q1 - (1.5 * iqr)
        
        print(f"Column: {colname}")
        print(f"Lower Extreme: {lower_extreme}")
        print(f"Upper Extreme: {upper_extreme}")
        print(f"Q1: {q1}")
        print(f"Q3: {q3}")
        print("------------------------")
        
        # Replace outliers
        Outlier_replacement(data, colname, upper_extreme, lower_extreme)

def Outlier_replacement(data, colname, upper_extreme, lower_extreme):
    data.loc[data[colname] > upper_extreme, colname] = upper_extreme
    data.loc[data[colname] < lower_extreme, colname] = lower_extreme

# Create a sample dataset
data = pd.DataFrame({
    'LB','AC','FM','UC','DL','DS','DP','ASTV','MSTV','ALTV','MLTV','Width','Tendency','NSP'
})

# Detect and replace outliers
Outlier_detection(df, ['LB', 'AC', 'FM', 'UC', 'DL', 'DS', 'DP', 'ASTV', 'MSTV', 'ALTV', 'MLTV', 'Width', 'Tendency', 'NSP'])
```
